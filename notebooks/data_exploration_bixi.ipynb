```python
#create a function to compare two years and find new bixi stations 
def compare_stations (df1 , df2 , col , dif=True):
    #compares two DataFrames and find the rows that are different
    #IMPUT:
    #df1 and df2: Data Frame to be compared
    #col : string - name of the column to be compared
    #new: boolean to select between the row that are diferent or the row that are equivalent
    #OUTPUT: 
    #Data Frame with the difernt or equivalent rows, depending on dif value

    df_merge = df1.merge(df2 , how='outer' , on=col , indicator=True)
    if dif:
        df_comp = df_merge[df_merge['_merge'] != 'both']
        
    else:
        df_comp = df_merge[df_merge['_merge'] == 'both']
    
    df_comp = df_comp.dropna(axis=1, how='all')
    df_comp = df_comp.drop(columns=["_merge"])
    return df_comp
```
